```python


# Step 1: Rename columns
new_col_names = []

for i in range(len(df.columns)):
    if 1 <= i <= 10:
        new_col_names.append(f'D{i}')  # i=2 â†’ D1, i=3 â†’ D2, ...
    elif 11 <= i <= 12:
        new_col_names.append(f'S{i-10}')  # i=6 â†’ S1, i=7 â†’ S2
    elif i >= 13:
        new_col_names.append(f'C{i-12}')  # i=8 â†’ C1, i=9 â†’ C2, ...
    else:
        new_col_names.append(df.columns[i])  # keep original for others (e.g., 0, 1)

# Step 2: Assign new column names
df.columns = new_col_names
```

In the pharmaceutical supply chain, an efficient routing and distribution model is essential to ensure timely
delivery of medicines. This study examines a multi-depot, multi-satellite, multi-compartment vehicle routing
problem in the context of medicine distribution in the districts of Mayurbhanj and Bhadrak, Odisha. The case
study focuses on the logistics of distributing 10 essential medicines from 10 different depots (distributors) to
two district-level intermediary points (Mayurbhanj and Bhadrak), which then distribute the medicines to 41
local hospitals, known as DHH (District Headquarters Hospitals) and CHCs (Community Health Centres).
The medicine distribution involves three hierarchical levels:
1. Depot Level (Distributors): The medicine is sourced from different distribution depots located across
five cities: Nagpur, Cuttack, Kolkata, Bhubaneshwar, and Noida at different locations. These depots
handle specific medicines:
â€¢ Nagpur: Vitamin B
â€¢ Cuttack: Diclofenac, Ofloxacin, Paracetamol
â€¢ Kolkata: Amlodipine
â€¢ Bhubaneshwar: Azithromycin, Metformin, Metronidazole
â€¢ Noida: Fluconazole, Levo Citrizin

2. Intermediate level (Satellites or district hospitals): Two district-level hospitals serve as interme-
diate distribution points:

â€¢ Mayurbhanj
â€¢ Bhadrak
These points receive medicines from the distributors and act as storage and distribution hubs for local
hospitals.
3. Local Level (Hospitals): The 41 local hospitals (DHHs and CHCs) under the Mayurbhanj and Bhadrak
districts serve as the final destinations for distributing the medicines.
The distribution network is multi-layered with different medicines sourced from specific depots and channeled
to intermediate hospitals (satellites) for onward distribution to local hospitals. The network flow for each
medicine can be described as follows:
1. Vitamin B is sourced from Nagpur and distributed to Mayurbhanj and Bhadrak via the Mayurbhanj
satellite hospital.
2. Diclofenac is sourced from Cuttack and distributed to Mayurbhanj and Bhadrak through both satellite
hospitals.
3. Amlodipin is sourced from Kolkata and flows to both satellite hospitals before being distributed to the
local hospitals.
4. Azithromycin is sourced from Bhubaneshwar, serving both Mayurbhanj and Bhadrak.
5. Fluconazole is sourced from Noida and distributed to both intermediate points.
6. Levo Citrizin comes from Noida, allowing for flexibility and redundancy in supply routes.
7. Metformin is sourced from Bhubaneshwar and distributed to the two intermediate points.
8. Metronidazole is also supplied from Bhubaneshwar, which acts as the main depot.
9. Ofloxacin is sourced from Cuttack and distributed to the satellite points.
10. Paracetamol is sourced from Cuttack and follows the same flow as Diclofenac and Ofloxacin.

To optimize the delivery process, the supply chain utilizes multi-compartment vehicles. Each compartment
is dedicated to specific medicine types, ensuring that temperature-sensitive or incompatible drugs are stored
appropriately and delivered without contamination. The use of multi-compartment vehicles also reduces the
number of trips required, leading to cost efficiency and a reduction in lead times.
After the medicines reach the satellite hospitals in Mayurbhanj and Bhadrak, the next step is to distribute
them to the 41 local hospitals. The distribution plan at this stage needs to account for delivery frequency
depending on the criticality and consumption rate of the medicines, deliveries are planned weekly, biweekly,
or monthly to avoid overstocking or shortages, and optimized routing so that local hospitals are distributed
across different blocks in Mayurbhanj and Bhadrak districts. Efficient routing strategies are designed to
cover multiple hospitals in a single trip, with routes planned to minimize travel distances and times while
ensuring timely delivery of medicines.
